```python
import random
import pandas as pd
```

```python
ACTIONS = [
    "move",
    "shoot",
    "hide",
    "reload",
    "explore",
    "loot",
    "fight",
    "die",
    "win",
    "quit"
]
```

```python
PLAYER_TYPES = {
    "aggressive": ["move", "shoot", "fight", "reload", "die", "win"],
    "explorer": ["move", "explore", "loot", "explore", "fight", "win"],
    "casual": ["move", "explore", "loot", "quit"],
    "quitter": ["move", "die", "quit"]
}
```

```python
def generate_player_logs(player_id, behavior , max_steps = 20):
    actions = PLAYER_TYPES[behavior]
    num_steps = random.randint(5, max_steps)

    log =[]
    for t in range(1,num_steps +1):
        action = random.choice(actions)
        log.append({
            "player_id": player_id,
            "timestamp": t,
            "action": action
        })

        if action in ["quit","die", "win"]:
            break

    return log

    
```

```python
all_logs = []

num_players = 100

for player_id in range(1, num_players + 1):
    behavior = random.choice(list(PLAYER_TYPES.keys()))
    player_log = generate_player_logs(player_id, behavior)
    all_logs.extend(player_log)

df = pd.DataFrame(all_logs)
print(df.head())
print(df.columns)

```

```python
df.to_csv("../data/raw/players.csv", index=False)
print("Saved to data/raw/players.csv")
```
